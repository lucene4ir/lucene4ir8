```python
import matplotlib.pyplot as plt
# CVS Line Format
# Header:
# 1 - Key Group  - 8 values
# corpus - indexType - qryFilter - qryCount - model - maxResults - RetrievabilityB - RetrievalCoefficient(B, mu, c)
# 2- ret File Group - 3 values
# G, ZeroRValuesCount, SumR
# 3 - res File - 3 values
# resLineCount - resDocCount - limitedQueries < max
# 4 - Performance - 3 values
# MAP, pbref, P10


# Column Indexes
# Key Group Index
modelCoefficientIndex = 8
# Retrievability Index
GIndex = modelCoefficientIndex + 1
sumRIndex = GIndex + 2

# Performance Group Index
mapIndex = 15
bprefIndex = mapIndex + 1
p10Index = mapIndex + 2

# Retrievability Dummy Index ( Differentiate Between Gravity and Cumulative )

gravityGIndex = 20
cumulativeGIndex = gravityGIndex + 1
allValue = 99
maxResults = ""
C = 0

sep = ","

def plotRange(xValues,yValues,indexType):
    plt.plot(xValues, yValues, label= indexType ,
             marker=getMarker(indexType),
             markerSize=8
             )

def findExSet (aKey,path):
    f = open(path,'r')
    found = False
    result = []
    for line in f :
        if line.startswith(aKey):
            found = True
            result.append(line)
        elif found:
            break
    return result

def getChartType(y):
    switcher = {
        cumulativeGIndex: "G - Cumulative",
        gravityGIndex : "G - Gravity - B 0.5",
        mapIndex: "Map Values",
        p10Index : "P10 Values",
        bprefIndex : "BPref Values",
        sumRIndex : "Total Retrievability Mass"
    }
    return switcher.get(y,"nothing")

def getMarker (i):
    switcher = {
        'UnigramIndex': "^", # Upper Triangle
        'BigramIndex': "o", # Circle
        'CombinedIndex': "X", # X Filled
        'FieldedIndex': "*"  # Star
    }
    return switcher.get(i, "nothing")

def getIndexType(i):
    i = i.upper()
    switcher = {
        'U':"UnigramIndex",
        'B': "BigramIndex",
        'C' : "CombinedIndex",
        'F': "FieldedIndex"
    }
    return switcher.get(i,"")

def getModelParameter (model):
    switcher = {
            "BM25":"B",
             "LMD": "mu",
             'PL2' : "c",
         }
    return switcher.get(model,"nothing")


def plotShow (x , y,corpus,model):
    global c
    if (x == modelCoefficientIndex):
        xTitle = getModelParameter(model)
    else:
        xTitle = getChartType(x)
    yTitle = getChartType(y)
    pltTitle = "%s Index - %s - (%s-%s) \n Chart C = %s" % (corpus , model, xTitle, yTitle,c)
    plt.ylabel(yTitle)
    plt.xlabel(xTitle)
    plt.title(pltTitle)

    #plt.xticks(xTicks)
    #plt.legend(["Unigram Index" , "BigramIndex" , "Combined Index" , "Fielded Index"])
    plt.legend()
    #,loc = "upper center", bbox_to_anchor = (0, -0.07)
    plt.show()

def getColValue (line , colIndex):
    parts = line.split(",", colIndex)
    value = parts[colIndex - 1]
    value = float(value)
    return value

def getColumn (xColIndex, yColIndex, indexType , path , corpus , model):
    resultY = []
    resultX = []
    inKey = inputKey(yColIndex, indexType ,  corpus , model)
    exSet = findExSet(inKey,path)
    if (yColIndex == gravityGIndex or yColIndex == cumulativeGIndex):
        yColIndex = GIndex
    if (xColIndex == gravityGIndex or xColIndex == cumulativeGIndex):
        xColIndex = GIndex
    for line in exSet:
        if (yColIndex != ""):
           value =  getColValue(line, yColIndex)
           resultY.append(value)
        if (xColIndex != ""):
            value = getColValue(line, xColIndex)
            resultX.append(value)

    return resultX , resultY


def colType (colIndex):
    if (colIndex == gravityGIndex or
            colIndex == cumulativeGIndex or
            colIndex == sumRIndex):
        result = 100
    else:
        result = 1000
    return result

def inputKey(colIndex, indexType ,  corpus , model ):
    # Qry Filter - Currently Constant: combinedQuery
    global c
    qryFilter = "combinedQuery"
    b = "0"
    if (colIndex==gravityGIndex):
        b = "0.5"

    if (colType(colIndex) == 100):
        # Query Count : 50 - 300K
        qryCount = "300K"
        # maxResults : 100 - 1000
        c = "100"

    else:
        # Query Count : 50 - 300K
        qryCount = "50"
        # maxResults : 100 - 1000
        c = "1000"


    key = sep.join([corpus, indexType, qryFilter, qryCount, model, c, b])
    key += sep
    return key

def csvReader(allX,allY,indexType,path,corpus , model):
    # Check if All Models
    if (corpus == allValue):
        allCorpuses = ["Aquaint","Core17","WAPO"]
    else:
        allCorpuses = [corpus]
    if (model == allValue):
        allModels = ["BM25","LMD","PL2"]
    else:
        allModels = [model]

    # Check if All Indexes
    if (indexType == allValue):
        allIndexes = "U B C F".split(" ")
    elif (indexType == ""):
        allIndexes = []
    else:
        allIndexes = [indexType]

    if (not isinstance(allY,list)):
        allY = [allY]
    if (not isinstance(allX,list)):
        allX = [allX]
    for corpus in allCorpuses:
        for model in allModels :
            for xIndex in allX:
                for yIndex in allY:
                    for i in allIndexes:
                        sameColType = xIndex == modelCoefficientIndex or colType(xIndex) == colType(yIndex)
                        currentIndexType = getIndexType(i)
                        if (sameColType):
                            xValues, yValues = getColumn(xIndex, yIndex, currentIndexType, path, corpus , model)
                        else:
                            temp, xValues = getColumn("", xIndex, currentIndexType, path, corpus , model)
                            temp, yValues = getColumn("", yIndex, currentIndexType, path, corpus , model)
                        plotRange(xValues, yValues, currentIndexType)
                    plotShow(xIndex, yIndex, corpus, model)
                    print ("Printed " + str(xIndex) + " " + str(yIndex))

def csvRetSplitter():
    newPath = "C:/Users/kkb19103/Desktop/CSV/ALLoutNew.csv"
    tempLines=[]
    fLines = []
    f = open(path,'r')
    hLine = f.readline()
    for line in f:
        parts = line.split(sep,6)
        c = parts[5]
        if (c=="1000"):
            for l in tempLines:
                fLines.append(l)
            tempLines = []
        else:
            gravity = line.split (sep , 17)[17]
            key = line.split (sep , 11)[11]
            key = line.replace(key,"")
            tempLine = key + gravity.replace("\n"," , , , , , , \n")
            tempLines.append(tempLine)
        fLines.append(line)
    f.close()
    f = open(newPath,'w')
    f.write(hLine)
    for line in fLines:
        f.write(line)
    f.close()

def setMainParameters():
    # Input
    # 10 - 100 - 1000

    # BM25 - PL2 - LMD - allValue
    model = "PL2"
    # Aquaint - Core17 - allValue
    corpus = "Core17"


    # Index Type :
    # U (UnigramIndex) -
    # B (BigramIndex) -
    # C (CombinedIndex) -
    # F (FieldedIndex)
    indexType = allValue
    path = r"C:\Users\kkb19103\Desktop\My Files 07-08-2019\BiasMeasurementExperiments\out.csv"
    x = [modelCoefficientIndex]
    y = [bprefIndex]

    return x , y , indexType , path , corpus , model

if __name__ == '__main__':

    #csvRetSplitter()
    x , y , indexType , path , corpus , model = setMainParameters()
    csvReader(x,y,indexType,path,corpus, model)

```
